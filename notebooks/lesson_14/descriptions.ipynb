````python
PARAMETERS = """
–°–≤–æ–π—Å—Ç–≤–æ: 
–ó–Ω–∞—á–µ–Ω–∏–µ
-----
y_–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ–û–ø–∏—Å–∞–Ω–∏–µ–ù–æ–º–µ–Ω–∫–ª–∞—Ç—É—Ä—ã–°–∫–∏–¥–∫–∏: 
–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—É–º–º–∞ –∑–∞–∫–∞–∑–∞ 1000 –≥—Ä–Ω.
–æ—Ç 2000 –≥—Ä–Ω. —Å–∫–∏–¥–∫–∞ 3%
 

-----
y_–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ–û–ø–∏—Å–∞–Ω–∏–µ–†–æ–∑–Ω–∏—Ü–∞: 
–ñ—ñ–Ω–æ—á–∞ —Ñ—É—Ç–±–æ–ª–∫–∞ –≤ —Ä—É–±—á–∏–∫

–í–ò–†–û–ë–ù–ò–ö: –¢—É—Ä–µ—á—á–∏–Ω–∞
–†–û–ó–ú–Ü–†: 56(S), 60(M), 70(XL), 80(3XL)
–Ø–ö–Ü–°–¢–¨: 100% –±–∞–≤–æ–≤–Ω–∞
-----
y_–ö–∞—Ç–µ–≥–æ—Ä–∏—èKASTA: 

-----
y_–ö–∞—Ç–µ–≥–æ—Ä–∏—è–†–æ–∑–Ω–∏—Ü–∞: 

-----
y_–†–∞–∑–¥–µ–ªKASTA: 

-----
y_–£—Å–ª–æ–≤–∏—è–ü—Ä–æ–¥–∞–∂–∏: 
–î–æ—Å—Ç–∞–≤–∫–∞ –∑–∞–º–æ–≤–ª–µ–Ω—å –∑–¥—ñ–π—Å–Ω—é—î—Ç—å—Å—è –∑—Ä—É—á–Ω–∏–º –¥–ª—è –∫–ª—ñ—î–Ω—Ç–∞ —Å–ø–æ—Å–æ–±–æ–º.
–ú–∏ —Å–ø—ñ–≤–ø—Ä–∞—Ü—é—î–º–æ –∑ —Ç–∞–∫–∏–º–∏ —Å–ª—É–∂–±–∞–º–∏, —è–∫ –ù–æ–≤–∞ –ü–æ—à—Ç–∞ —Ç–∞ –£–∫—Ä–ü–æ—à—Ç–∞.
–í–∞—Ä—Ç—ñ—Å—Ç—å –¥–æ—Å—Ç–∞–≤–∫–∏ –∑–∞–ª–µ–∂–∏—Ç—å –≤—ñ–¥ –≤–∞–≥–∏ –ø–æ—Å–∏–ª–∫–∏ —Ç–∞ —Ä–æ–∑—Ä–∞—Ö–æ–≤—É—î—Ç—å—Å—è —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ—é –∫
-----
–ê—Ä—Ç–∏–∫—É–ª: 
—ç–∑–≥2152
-----
–í–∏–¥–ù–æ–º–µ–Ω–∫–ª–∞—Ç—É—Ä—ã: 
—Ç–æ–≤–∞—Ä
-----
–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ–û–ø–∏—Å–∞–Ω–∏–µ–ù–æ–º–µ–Ω–∫–ª–∞—Ç—É—Ä—ã: 

–§—É—Ç–±–æ–ª–∫–∞ –∂–µ–Ω—Å–∫–∞—è –≠–∑–≥–∏ (–ª–∞–ø—à–∞) 2152
–ü–†–û–ò–ó–í–û–î–ò–¢–ï–õ–¨: Ezgi, –¢—É—Ä—Ü–∏—è.
–†–ê–ó–ú–ï–†: 56(S), 60(M), 70(XL), 80(3XL),
–¶–í–ï–¢: –±–µ–ª—ã–π, —á–µ—Ä–Ω—ã–π, –±–µ–∂–µ–≤—ã–π.
–ö–ê–ß–ï–°–¢–í–û: 100% —Ö–ª–æ–ø–æ–∫.
–¢–æ–≤–∞—Ä –ø—Ä–æ–¥–∞–µ—Ç—Å—è —É–ø–∞–∫–æ–≤–∫–∞–º–∏. –£–ø–∞–∫–æ–≤–∫–∞ —Å–æ–¥–µ—Ä–∂
-----
–ö–∞—Ç–µ–≥–æ—Ä–∏—èPromUa: 
–ú–∞–π–∫–∏ –∏ —Ñ—É—Ç–±–æ–ª–∫–∏
-----
–ö–æ–¥: 
12795
-----
–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π: 

-----
–ù–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ: 
–≠–∑–≥–∏ —Ñ—É—Ç–±–æ–ª–∫–∞ –ª–∞–ø—à–∞ (–±–µ–∂–µ–≤—ã–π,–±–µ–ª—ã–π,—á–µ—Ä–Ω—ã–π) (—É–ø.6—à—Ç*2=12—à—Ç.) 2152
-----
–ù–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ–ü–æ–ª–Ω–æ–µ: 
–§—É—Ç–±–æ–ª–∫–∞ –∂–µ–Ω—Å–∫–∞—è Ezgi (–ª–∞–ø—à–∞) 2152
-----
–ù–æ–º–µ–Ω–∫–ª–∞—Ç—É—Ä–Ω–∞—è–ì—Ä—É–ø–ø–∞: 
–§—É—Ç–±–æ–ª–∫–∏, –º–∞–π–∫–∏
-----
–¶–µ–Ω–æ–≤–∞—è–ì—Ä—É–ø–ø–∞: 
–≠–∑–≥–∏ –ª–∞–ø—à–∞
-----

"""
````
